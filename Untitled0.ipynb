```
import cv2, numpy as np, matplotlib.pyplot as plt
from scipy import fftpack
%matplotlib inline
img = cv2.imread('drive/My Drive/Colab Notebooks/Alz_dset/Test/MildDemented/26 (19).jpg',0)
img_log = (np.log(1+img)/np.log(1+np.max(img)))*255
img_log = np.array(img_log, dtype = np.uint8)
#display(Image('drive/My Drive/Colab Notebooks/mri_test.jpg'))
plt.imshow(cv2.imread('drive/My Drive/Colab Notebooks/Screenshot (20).png'))
plt.title('Transformation graphs')
plt.show()
plt.imshow(img)
plt.title('Original')
plt.show()
axes = plt.gca()
axes.set_ylim([0,1000])
hist = plt.hist(img.ravel(),256,[0,256])
plt.title('original histogram')
plt.show()

#FFT start

im_fft = fftpack.fft2(img)

keep_fraction = 0.1
im_fft2 = im_fft.copy()
print(im_fft2.shape)
r, c = im_fft2.shape
#im_fft2[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
#im_fft2[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0

im_new = fftpack.ifft2(im_fft2).real
plt.figure()
plt.imshow(im_new, plt.cm.gray)
plt.title('Filtered of original')
plt.show()

#FFT end

#Log transformed 

plt.imshow(img_log)
plt.title('Log transformation')
plt.show()
axes = plt.gca()
axes.set_ylim([0,1000])
hist = plt.hist(img.ravel(),256,[0,256])
plt.title('Log histogram')
plt.show()

#FFT start


im_log_fft = fftpack.fft2(img_log)

keep_fraction = 0.1
im_log_fft2 = im_log_fft.copy()
r, c = im_log_fft2.shape
im_log_fft2[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
im_log_fft2[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0

im_log_new = fftpack.ifft2(im_log_fft2).real
plt.imshow(im_log_new, plt.cm.gray)
plt.title('Filtered of log of original')
plt.show()

#FFT end
```
